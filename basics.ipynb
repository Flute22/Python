# Python 
    * Python is a high-level, interpreted programming language designed fro readability and rapid development - think of it as "executable pseudocode" that prioritizes human understanding over machine efficiency. 

    * Python was created to solve a specific problem: most languages force you to think like the computer instead of thinking like a human. Python flips this - you write what you mean, and Python handles the translation to machine code. 

## print() function -:

Key behaviors:

* Automatically adds a newline at the end (unless you change it)

* Separates multiple arguments with a space by default

## Variables
    Are names that store references to objects in memory.


    ===> Naming Conventions: 
        1. Must start with letter or underscores: name, _private. 
        2. Can contain letters, numbers, underscores: user_1, myVar2. 
        3. Case-Sensitive: userName, username, UserName.
        4. Can't use reserved keywords: class, def, for, while, etc. 
        5. Use meaningful names: user_name, user_age, user_email etc. 
    
    ===> Creating & Assigning: 
        * Simple Assignment: 
            name = "Khushal Sinhmar"
            age = 22
        
        * Multiple Assignment (tuple unpacking): 
            name, age = "Khushal Sinhmar", 22

        * Same values to multiple variables: 
            a = b = c = 10
        
        * Swapping values: 
            a, b = b, a
        
        * Chain assignment with different values: 
            x = (y:=20) + 5

## Operators 
    Operators are symbols that perform operations on values or variables. 

```python
# ==> Arithmetic Operators: Basic mathematics operations.

a = 10
b = 3

a + b    # 13  (Addition)
a - b    # 7   (Subtraction)
a * b    # 30  (Multiplication)
a / b    # 3.333... (Division - always returns float)
a // b   # 3   (Floor division - rounds down)
a % b    # 1   (Modulus - remainder)
a ** b   # 1000 (Exponentiation - power)



# ==> Comparison Operators: Compare values, reutrn True or False

a = 10
b = 5

a == b   # False (Equal to)
a != b   # True  (Not equal to)
a > b    # True  (Greater than)
a < b    # False (Less than)
a >= b   # True  (Greater than or equal)
a <= b   # False (Less than or equal)



# ==> Logical Operators: Combine boolean conditions.

a = True
b = False

a and b  # False (Both must be True)
a or b   # True  (At least one must be True)
not a    # False (Inverts the value)

# Practical example
age = 18
if age >= 18 and age < 65:
    print("Working age")


# ==> Assignment Operators: Assign & modify values.

x = 10        # Basic assignment

x += 5        # x = x + 5  → 15
x -= 3        # x = x - 3  → 12
x *= 2        # x = x * 2  → 24
x /= 4        # x = x / 4  → 6.0
x //= 2       # x = x // 2 → 3.0
x %= 2        # x = x % 2  → 1.0
x **= 3       # x = x ** 3 → 1.0



# ==> Identity Operator: Check if variable point to the same object in memory. 

a = [1, 2, 3]
b = a
c = [1, 2, 3]

a is b      # True  (same object)
a is c      # False (different objects, same values)
a is not c  # True

# Commonly used with None
x = None
if x is None:
    print("x is None")



# ==> Mermbership Operators: Check if value exists in the sequence. 

fruits = ["apple", "banana", "cherry"]

"apple" in fruits      # True
"grape" in fruits      # False
"grape" not in fruits  # True

# Works with strings too
"hello" in "hello world"  # True



# ==> Bitwise Operators: Operate on binary representation.

a = 5   # Binary: 0101
b = 3   # Binary: 0011

a & b   # 1  (AND: 0001)
a | b   # 7  (OR: 0111)
a ^ b   # 6  (XOR: 0110)
~a      # -6 (NOT: inverts bits)
a << 1  # 10 (Left shift: 1010)
a >> 1  # 2  (Right shift: 0010)



# ==> Walrus Operator: Assign and return values in one expression. 

# Without walrus
data = input("Enter name: ")
if len(data) > 3:
    # print(f"{data} is valid")
    pass

# With walrus
if (n := len(input("Enter name: "))) > 3:
    # print(f"Length {n} is valid")
    pass

# Useful in loops
while (line := input(">>> ")) != "quit":
    # print(f"You said: {line}")
    pass



# ==> Ternary Operator: Compact if-else in one line.

# Without walrus
data = input("Enter name: ")
if len(data) > 3:
    # print(f"{data} is valid")
    pass

# With walrus
if (n := len(input("Enter name: "))) > 3:
    # print(f"Length {n} is valid")
    pass

# Useful in loops
while (line := input(">>> ")) != "quit":
    # print(f"You said: {line}")
    pass



# Find the sum of 3 digit number enterd  by the user
digit = int(input("Enter 3 digit number: "))

for i in range(digit):
    last_digit = digit % 10
    sum += last_digit
    digit = digit // 10

sum

```

## If-Eles Conditions
    It make decisions in code - it runs different code based on whether a condition is True & False.

```python
email = input("Enter email: ")
password = input("enter password: ")

if email == "khushalsinhmar@gmail.com" and password == "kabir": 
    print("Welcome")
elif email == "khushalsinhmar@gmail.com" and password != "kabir": 
    print("Incorrect password")
    password = input("Enter password again: ")
    
    if password == "kabir": 
        print("Welcome finally")
    else: 
        print('Again not correct')
else: 
    print("Not Correct")
```

```python
# Simple calculator
fnum = int(input("Enter first number: "))
snum = int(input("Enter second number: "))
operator = input("Enter you operator: ")

if operator == '+': 
    print(fnum + snum)
elif operator == '-': 
    print(fnum - snum)
elif operator == '*': 
    print(fnum * snum)
elif operator == '/': 
    print(fnum / snum)
elif operator == '%': 
    print(fnum % snum)
elif operator == '//': 
    print(fnum // snum)
else: 
    print("Invalid Operator!")
```

```python
# Number gusseing game
import random
random_number = random.randint(1, 100)
print(random_number)
user_guess = int(input("Guess a number: "))
count = 0 

while user_guess != random_number: 
    if user_guess < random_number: 
        print("User is smaller than computer number!")
    elif user_guess > random_number: 
        print("User is higer than computer number!")
    else: 
        print("Invalid Guess!")

    user_guess = int(input("Guess a number: "))
    count += 1
else: 
    print("You gess it correct!")
    print(f'How many guess: {count}')
```

### Nested Loops

```python
# Print expression 
n = int(input("Enter a number: "))

for i in range(1, n+1): 
    for j in range(1, n+1): 
        print(i, j)
```

```python
# Pattern-1

n = int(input("Enter a number: "))

for i in range(1, n+1): 
    for j in range(i): 
        print('* ', end="")
    print()
```

```python
n = int(input("Enter a number: "))

for i in range(1, n+1): 
    for j in range(1, i+1): 
        print(j, end='')
    for k in range(i-1, 0, -1):
        print(k, end='')
    print()
```

### Loop Control Statements
    Control statements change how loops execute. 

    * break -: Exit the loop immediately.  
    
    * continue -: Skip the rest of this iteration and go to the next one. 
    
    * pass -: Do nothing (placeholder for empty blocks).

```python
# Prime number finder
s_range = int(input("Enter strating range: "))
e_range = int(input("Enter the ending range: "))

for i in range(s_range, e_range+1): 
    for j in range(2, i): 
        if i % j == 0: 
            break 
    else: 
        print(i)
```
