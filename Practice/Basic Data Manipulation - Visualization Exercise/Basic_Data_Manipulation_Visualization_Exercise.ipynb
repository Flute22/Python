### 50. Write a Python program to convert a string to datetime.
Sample String : Jul 1 2014 2:43PM 

Expected Output : 2014-07-01 14:43:00

```python
# Step 1: Define a simple date string 
date_text = "Jul 1 2014 2:43PM"

# Step 2: Define the exact matching blueprint format
# %Y = 4-digit year, %m = 2-digit month, %d = 2-digit day
# %H = 24-hour clock hour, %M = minute, %S = second
str_format = "%b %d %Y %H:%M%p"

# Step 3: Parse the string into datetime object
parsed_datetime = datetime.strptime(date_text, str_format)

# Step 4: View a clean human-readable representation using strftime 
clean_output = parsed_datetime.strftime("%Y-%B-%d %I:%M")

clean_output 
```

```python
# Step 1: Capture the current calender year from the host machine runtime. 
# This prevents hardcoding boundaries (e.g., automatically evaluates to 2026)
current_year = pd.Timestamp.now().year

# Step 2: Identify invalid dates using an index condition
safe_timeline_condition = data.index.year <= current_year 

# Step 3: Apply the vectorized shift exclusively to dates. 
# If True ( <= 2026 ): Keep value. If False ( > 2026 ): Subtract 100 years.
corrected_index = data.index.where(
    safe_timeline_condition,
    data.index - pd.DateOffset(years=100)
)

# Step 4: Assign the corrected index to the DataFrame
data.index = corrected_index

data.head()
```
