### `Problem 1`: Write a program that will give you in hand monthly salary after deduction on CTC - HRA(10%), DA(5%), PF(3%) and taxes deduction as below:

> Salary(Lakhs) : Tax(%)

*   Below 5 : 0%
*   5-10 : 10%
*   10-20 : 20%
*   aboove 20 : 30%

````python
def salary_deduction_on_ctc(ctc):
    """
    This function tasks ctc as parameter and deduct hra, da, pf and tax (if it come under the condition). 
    Eventually it return the monthly inhand salary.
    """
    
    hra = (10/100) * ctc
    da = (5/100) * ctc
    pf = (3/100) * ctc
    tax = 0

    ctc_in_lakhs = ctc / 100000

    if ctc_in_lakhs < 5: 
        tax = 0
    elif 5 <= ctc_in_lakhs <= 10:
        tax = (10/100) * ctc
    elif 10 <= ctc_in_lakhs <= 20:
        tax = (20/100) * ctc
    elif ctc_in_lakhs >= 20:
        tax = (30/100) * ctc

    total_deduction = hra + da + pf + tax

    annual_inhand = ctc - total_deduction

    monthly_inhand = annual_inhand / 12

    return monthly_inhand


ctc = 1000000
salary_deduction_on_ctc(ctc)
````

### `Problem 2`: Write a program that take a user input of three angles and will find out whether it can form a triangle or not.

````python
def valid_triangle(a, b , c): 
    """
    This function takes 3 inputs as parameter. First is A, second is B, and third one C (3 angles of triangle).
    Eventually it return a string, which tell's it can form a valid triangle or not?

    Note -> A valid triangle = If sum of all angels is equal to 180 and all angles are greater than 0.
    """

    if a + b + c == 180 and all(x > 0 for x in [a, b, c]): 
        return "It form valid triangle!"
    else: 
        return "It doesn't form triangle!"


a = int(input("Enter the first angle: "))
b = int(input("Enter the second angle: "))
c = int(input("Enter the third angle: "))

print(valid_triangle(a, b, c))
````

### `Problem 3`: Write a program that will take user input of cost price and selling price and determines whether its a loss or a profit.

````python
def is_profit(cp, sp): 
    """
    This function takes two input as parameter. 
    There return you are in profit or loose or neither.
    """

    if sp > cp: 
        print(f'You are in PROFIT! \nProfit: {sp - cp}')
    elif sp < cp: 
        print(f'You are in loose! \nLoss: {cp - sp}')
    else: 
        print(f'You earn nothing not in loss not in profit!')

cost_price = int(input("Enter a cost price: "))
selling_price = int(input("Enter a selling price: "))

is_profit(cost_price, selling_price)
````

### `Problem 4`: Write a menu-driven program -
1. cm to ft
2. km to miles
3. USD to INR
4. exit

````python
def menu(choice): 
    match choice: 
        case 1: 
            print("---- Converting CM to FT ----")
            value = int(input("Enter value: "))
            return round(value / 30.48, 2)

        case 2: 
            print("---- Converting KM to Miles ----")
            value = int(input("Enter value: "))
            return round(value * 0.621371, 2)
        case 3: 
            print("---- Converting USD to INR ----")
            value = int(input("Enter value: "))
            return round(value * 92, 2)
        case 4: 
            exit()
        case _:
            return "Invalid Error"
        
choice = int(input("Choose Case: "))

print(menu(choice))
````
