### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
import math

# 2D Dimension
def euclidean_distance_2d(coordinate1, coordinate2):
    """
        First this function take input as 2 coordinates.
        Then find the euclidean distance between those two coordinates.
        
        ---> Note: This function return output for 2D dimensions.
    """

    return math.sqrt((coordinate2[0] - coordinate1[0])**2 + (coordinate2[1] - coordinate1[1])**2)    
    

euclidean_distance_2d((2, 1), (3, 4))
````

````python
# nD Dimensions
def euclidean_distance_nD(coordinate1, coordinate2): 
    """
        First this function take input as 2 coordinates.
        Then find the euclidean distance between those two coordinates.
        
        ---> Note: This function return output for nD dimensions.
    """

    if len(coordinate1) != len(coordinate2): 
        return "Error: Both points must have same dimensions!"

    return math.sqrt(sum((b - a)**2 for a, b in zip(coordinate1, coordinate2)))

euclidean_distance_nD((1, 5, 2, 6, 4, 8), (2, 34, 5, 2, 1, 2))
````

### Q5:- Write a program to find the simple interest when the value of principle,rate of interest and time period is provided by the user.


````python
def simple_interest(p, r, t): 
    """
    This simple interest function 3 input from user: principle, rate, and time
    Then return the Interest. 
    """

    return (p * r * t)/100


simple_interest(10000, 2, 2)
````

### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
def number_of_dogs_and_chickens(heads, legs): 
    """
    This function take two parameters: heads and legs. 
    Then return you how many are dogs and how many are chickens.

    --> I this function I use a Formula to solve this. In this formula we have to assume all are chicken means all have 2 legs. 
    """

    dogs = round((legs - (heads * 2)) / 2) # So, we are finding how many dogs, thats why we divide here with 2 because dogs have 2 extra legs. 
    chickens = heads - dogs

    print(f'Dogs: {dogs} \nChickens: {chickens}')

number_of_dogs_and_chickens(8, 30)
````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

````python
# Solution with formula

def sum_of_squares(n): 
    """
    This function take input as number. 
    Then return the sum of squares till that number. 
    ---> It return the output using formula.
    """    

    return (n * ((n+1) * (2 * n + 1))) / 6

sum_of_squares(4)
````

````python
# Solution using loop

def sum_of_squares(n): 
    """
    This function take input as number. 
    Then return the sum of squares till that number. 
    ---> It return the output using loop.
    """  
    # sum = 0
    # for i in range(1, n+1):
    #     sum += i**2
    
    return sum(i**2 for i in range(1, n+1))

sum_of_squares(4)
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
def nth_term(a, b, t): 
    if b < a or b == a: 
        return "Error: You have to choose second term bigger then first term"

    return a + (t - 1) * (b - a)

fist_term = int(input("Enter first term: "))
second_term = int(input("Enter second term: "))
f_term = int(input("Which term do you want to find: "))

nth_term(fist_term, second_term, f_term)
````

### Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

````python
def sum_of_2_fractions(n1, d1, n2, d2):
    return ((n1 * d2) + (n2 * d1)) / (d1 * d2)

n1 = int(input("Enter numerator of fraction 1: "))
d1 = int(input("Enter denominator of fraction 1: "))

n2 = int(input("Enter numerator of fraction 2: "))
d2 = int(input("Enter denominator of fraction 2: "))

sum_of_2_fractions(n1, d1, n2, d2)
````

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
import math

def glasses_of_milk(tank_h, tank_l, tank_b, glass_h, glass_r): 
    volume_of_tank = tank_h * tank_l * tank_b
    volume_of_glass = math.pi * (glass_r**2) * glass_h

    return f'Number of glasses: {round(volume_of_tank / volume_of_glass)}'

glasses_of_milk(20, 20, 20, 3, 1)
````
