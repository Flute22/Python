### Pandas - Series

```python
import pandas as pd
import numpy as np
```

#### Creating Series using List

```python
# String Series
country = ['India', 'Pakistan', 'USA', 'Nepal', 'Japan', 'Dubai', 'Egypt']
pd.Series(country)
```

```python
# Integer Series
runs = [18, 19, 20, 21, 22]
pd.Series(runs)
```

```python
# Custom Index Series
marks = [50, 78, 90, 85, 93]
subjects = ['Hindi', 'History', 'English', 'Maths', 'Science']

pd.Series(marks, index=subjects)
```

```python
# Setting a name in Series
marks = [50, 78, 90, 85, 93]
subjects = ['Hindi', 'History', 'English', 'Maths', 'Science']

marks = pd.Series(marks, index=subjects, name='Marks')
marks
```

#### Creating Series using Dictionary

```python
marks_dict = {
    'maths': 98,
    'science': 80, 
    'English': 100, 
    'History': 89
}

dict_series = pd.Series(marks_dict)
dict_series
```
