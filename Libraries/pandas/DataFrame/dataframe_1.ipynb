### DataFrame Attributes and Methods

```python
# shape --> This attribute is used to quickly identified the dimensions of DataFrame. 
# It returns a tuple in the format (rows, columns).
print(movies.shape)
print(ipl.shape)
```

```python
# dtypes --> It returns a Series showing data types of each column. 
movies.dtypes
ipl.dtypes
```

```python
# columns --> It returns the column labels of the DataFrame.
movies.columns
ipl.columns
```

```python
# values --> It returns all the values from DataFrame.
movies.values
ipl.values
students.values
```

```python
# head() --> It return top 5 vales. Apart from that, you can manipulate the values according, how many values you want. 
movies.head()
ipl.head()
```

```python
# tail() --> It return lower 5 vales. Apart from that, you can manipulate the values according, how many values you want. 
movies.tail()
ipl.tail()
```

```python
# sample() --> Fetch a random value from dataframe. Additionally, you could manipulate the values accordingly, how many you want.
movies.sample(2)
```

```python
# info() --> It prints a concise summary of a DataFrame. 
movies.info()
```

```python
# describe() --> It generates descriptive statistics that summarize the central tendency, dispersion and shape of dataset's distribution.
movies.describe()
```

```python
# isnull() --> It returns True if it find null value. For more effective result you should use it with .sum() method. 
movies.isnull().sum()
```

```python
# duplicated() --> To find the duplicate values
students.duplicated().sum()
```

```python
# rname() --> It could change the name of column. However it will not change in original table or dataframe. 
students.rename(columns={'marks': 'percent', 'package': 'lpa'})
```

### Filtering a DataFrame

```python
ipl.head(1)
```

```python
# find all the final winners
mask = ipl['MatchNumber'] == 'Final'
new_df = ipl[mask]

new_df[['Season', 'WinningTeam']]
```

```python
# one-liner
ipl[ipl['MatchNumber'] == 'Final'][['Season', 'WinningTeam']]
```

```python
# How many super over finishes have occurred?
ipl[ipl['SuperOver'] == 'Y'].shape[0] 
```

```python
# How many matches has csk won in Kolkata
ipl[(ipl['City'] == 'Kolkata') & (ipl['WinningTeam'] == 'Chennai Super Kings')].shape[0]
```

```python
# toss winner in match winner in percentage
print(f'{(ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0]/ipl.shape[0])*100: .2f}')
```

```python
movies.head(1)
```

```python
# movies with rating higher than 8 and votes > 10000
movies[(movies['imdb_rating'] > 8) & (movies['imdb_votes'] > 10000)].shape[0]
```

```python
# action movies with rating higher than 7.5
mask1 = movies['genres'].str.split('|').apply(lambda x: 'Action' in x)
mask2 = movies['imdb_rating'] > 7.5

movies[mask1 & mask2].shape[0]
```

```python
mask1 = movies['genres'].str.contains('Action')
mask2 = movies['imdb_rating'] > 7.5

movies[mask1 & mask2].shape[0]
```
