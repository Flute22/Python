### Pandas - Series

### Some basic functionality

```python
import numpy as np
import pandas as pd
```

```python
countries = ["Japan", "Switzerland", "Norway", "Dubai", "China", "Finland", "Venezuela"]
pd.Series(countries)
```

```python
runs = [4, 21, 22, 18, 19, 20]
pd.Series(runs)
```

```python
# custom index
marks = [85, 43, 60, 78, 90, 91]
subjects = ["Maths", "Science", "Hindi", "English", "History", "Polity"]

marks_series = pd.Series(marks, index=subjects, name='my marks')
```

### Series Attributes

```python
# size
marks_series.size
```

```python
# dtype
marks_series.dtype 
```

```python
# name 
marks_series.name
```

```python
# is_unique
marks_series.is_unique
```

```python
# head -> It provide top 5 rows, In addition to that, if want more or less rows you can change it accordingly.
subs.head()
```

```python
vk.head(4)
```

```python
movies.head(10)
```

```python
# tail -> It provide last 5 rows, Apart from that, if you want more or less rows you can change it accordingly. 
vk.tail()
```

```python
vk.tail(10)
```

```python
# sample -> it give the one row randomly from all the rows. Additional if you want more or less rows you can change it accordingly.
movies.sample()
```
