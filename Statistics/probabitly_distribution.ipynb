```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
List = []

for i in range(10000): 
    dice1 = random.randint(1, 6)
    dice2 = random.randint(1, 6)
    List.append(dice1 + dice2)
```

```python
dice1_series = pd.Series(List)
dice1_series
```

```python
dice_prob = (dice1_series.value_counts() / dice1_series.value_counts().sum()).sort_index()

dice_prob
```

```python
dice_prob.plot(kind='bar')
```

```python
CDF = np.cumsum(dice_prob)
```

```python
CDF
```

```python
CDF.plot(kind='bar')
```
