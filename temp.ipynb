```python
def greet(name, age):
    return f'Hello {name}, You age is {age}'

name = input("What is your name: ")
age = input("What is your age: ")

print(greet(name, age))
```

```python
# Q3: Take two numbers and print their sum, difference, product, and division.

# Variable declaration 
num1 = 21
num2 = 22

# Calculating the sum
sum = num1 + num2
print(f"Sum: {sum}")

# Calculating the difference
diff = num1 - num2
print(f"Diff: {diff}")

# Calculating the product
prod = num1 * num2
print(f"Prod: {prod}")

# Calculating the division
div = num1 / num2
print(f"Div: {div:.2f}")
```

```python
# Q4: Check if a number is even or odd. 
user_num = int(input("Enter a number: "))

if user_num % 2 == 0: 
    print(f'{user_num} is even.')
else: 
    print(f'{user_num} is odd.')
```

```python
def is_even(num):
    if num % 2 == 0: 
        return f'{num} is even!'
    else: 
        return f'{num} is odd!'

num = int(input("Enter a num: "))
print(is_even(num))
```

```python
# Q5: Find the largest of three numbers.

# Variable declaration 
num1 = int(input("Enter a number"))
num2 = int(input("Enter a number"))
num3 = int(input("Enter a number"))

# Conditional Logic
if num1 >= num2:
    print(f'{num1} is greater')

elif num2 >= num3:
    print(f'{num2} is greater')

else:
    print(f'{num3} is greater')
```

```python
# Print numbers from 1 to 20 using a loop.
def one_to_twenty(): 
    for i in range(1, 21): 
        print(i)
        
one_to_twenty()
```

```python
# Count how many numbers between 1–100 are divisible by 7.
def divisible_by_7():
    for i in range(1,100):
        if i % 7 == 0:
            print(i)

divisible_by_7()    
```
